```python
import pandas as pd
import numpy as np
```

```python
authors = pd.DataFrame({'author_id' : [1, 2, 3],
                        'author_name': ['Тургенев', 'Чехов', 'Островский']})
authors
```

```python
books = pd.DataFrame({'author_id': [1, 1, 1, 2, 2, 3, 3],
                     'book_title': ['Отцы и дети', 'Рудин', 'Дворянское гнездо', 'Толстый и тонкий', 'Дама с собачкой', 'Гроза', 'Таланты и поклонники'],
                     'price': [450, 300, 350, 500, 450, 370, 290]})
books
```

```python
authors_prices = pd.merge(authors, books, on='author_id', how='inner')
authors_prices
```

```python
#top5 = authors_prices[authors_prices['price'] > 300]
top5 = authors_prices.sort_values('price')
top5.tail()
```

```python
authors_stat = pd.DataFrame({
    'max_price': authors_prices.groupby('author_name')['price'].max(),
    'mean_price': authors_prices.groupby('author_name')['price'].mean(),
    'min_price': authors_prices.groupby('author_name')['price'].min()})
authors_stat.reset_index()
```
