```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [3.5, 3.8, 4.2, 4.5, 5, 5.5, 7]
plt.plot(x, y)
plt.show()
```

```python
plt.scatter(x, y)
plt.show()
```

```python
t = np.linspace(0, 10, 51)
f = np.cos(t)
```

```python
plt.plot(t, f, color='green')
plt.title(' f(t)')
plt.xlabel(' t')
plt.ylabel(' f')
plt.axis([0, 9.5, -2.5, 2.5])
plt.show()
```

```python
x = np.linspace(-3, 3, 51)
y1 = x ** 2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)

ax1, ax2, ax3, ax4 = ax.flatten()
ax1.plot(x, y1)
ax1.set_title(" y1")
ax1.set_xlim([-5, 5])
ax2.plot(x, y2)
ax2.set_title(" y2")
ax3.plot(x, y3)
ax3.set_title(" y3")
ax4.plot(x, y4)
ax4.set_title(" y4")
```

```python
plt.style.use('fivethirtyeight')
fig, ax = plt.subplots(nrows=1, ncols=2)
fig.set_size_inches(10,4)
fig.subplots_adjust(wspace=0.3, hspace=0.3)

ax1, ax2 = ax.flatten()

ax1.hist(df[df['Class'] == 1]['V1'], density=True, bins=20)
ax1.set_title('class = 1')
ax1.alpha = 0.5

ax2.hist(df[df['Class'] == 0]['V1'], density=True, bins=20)
ax2.set_title('class = 0')
ax2.alpha = 0.5
```
