```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

import seaborn as sns
from scipy import stats
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
matplotlib.rcParams.update({'font.size': 14})  # размер шрифта на графиках
```

```python
from sklearn.datasets import load_boston

boston = load_boston()
data = boston['data']
feature_names = boston['feature_names']
target = boston['target']
```

```python
X = pd.DataFrame(data, columns=feature_names)
X.head()
```

```python
X.info()
```

```python
y = pd.DataFrame(target, columns=["price"])

y.info()
```

# LinearRegression

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.metrics import r2_score
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
r2 = r2_score(y_test, y_pred)
r2
```

# RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
n_estimators = 1000
max_depth = 12
random_state = 42
```

```python
rf = RandomForestRegressor(n_estimators=n_estimators, max_depth=max_depth, random_state=random_state)
rf.fit(X_train, y_train.values[:, 0])
y_pred = rf.predict(X_test)
r2 = r2_score(y_test, y_pred)
r2
```

RandomForest показывает лучшее качество

```python
help(RandomForestRegressor)
```

```python
print(f'Сумма {sum(rf.feature_importances_)}')
mvp_features = sorted(rf.feature_importances_, reverse=True)
print(f'Два наиболее важных признака {mvp_features[0]} {mvp_features[1]}')
```

```python
plt.barh(feature_names, rf.feature_importances_)

plt.xlabel("Вес признака")
plt.ylabel("Признак")

plt.show()
```

```python
df = pd.read_csv('creditcard.csv')
df
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import GridSearchCV
```

```python
df.value_counts(normalize=True)
```

```python
pd.options.display.max_columns = 100
df.info()
```

```python
df.head(10)
```

```python
X = df.iloc[:, :30]
y = df['Class']
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100, stratify=y)
```

```python
X_train.shape
```

```python
parameters = [{'n_estimators': [10, 15],
               'max_features': np.arange(3, 5),
               'max_depth': np.arange(4, 7)}]

grid = GridSearchCV(estimator=RandomForestClassifier(random_state=100), param_grid=parameters, scoring='roc_auc', cv=3)
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.predict_proba(X_test)
```

```python
y_pred_proba = grid.predict_proba(X_test)[:, 1]
```

```python
roc_auc_score(y_test, y_pred_proba)
```
